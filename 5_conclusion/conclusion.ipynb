```python
import numpy as np
import scanpy as sc
import matplotlib.pyplot as plt
```

We are justified in our choice of combining group 0 and 8

```python
data = sc.read("../4_diff_gene_analysis/diff_gene_data.h5ad")
```

```python
cluster_rename_dict = {
    "0": "Lower tubule PC",
    "1": "Principle Cell",
    "2": "Tubule bar shaped cell of initial segment",
    "3": "Stellate",
    "4": "Initial/transitional PC",
    "5": "Main segment PC",
    "6": "Adult Renal Stem Cell",
    "7": "Unidentified",
    "8": "Lower tubule PC",
    "9": "Tracheal cell",
    "10": "Principle cell of lower ureter OR Posterior midgut",
    "11": "Adult fat body"
}
```

```python
data.obs["leiden_renamed"] = data.obs["leiden"].map(cluster_rename_dict)
```

```python
sc.pl.umap(data, color="leiden_renamed", title="Leiden Clustering with Custom Names")
```

```python
sc.pl.umap(data, color="leiden_renamed", title="Leiden Clustering with Custom Names", show=False)
name = "../5_conclusion/UMAP_Leiden_graph_final.png"
plt.savefig(name, dpi=300, bbox_inches="tight")  #save it
```

We see 11 distinct cell types, that have been succesfully seperated, each with distinct marker genes, most of the cell types make sense for the malpighian tubules, except for the tracheal cells and the unidentified cluster.
